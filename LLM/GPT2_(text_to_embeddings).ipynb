```python
import torch
from transformers import GPT2Config, GPT2Model, GPT2Tokenizer

config = GPT2Config.from_pretrained('gpt2')
# Change the hyperparameters of model
# config.num_layers = 12
# config.hidden_size = 768
model = GPT2Model.from_pretrained('gpt2', config = config)
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

text = "Helo, how are you?"

# Tokenize the text
tokens = tokenizer.encode(text, add_special_tokens=True)
input_ids = torch.tensor([tokens])

# Disable gradient computation
with torch.no_grad():
    # Get the embeddings
    outputs = model(input_ids)
    #embeddings have dimension a*b*c
    # a: The number of sequences or input examples processed in parallel.
    # b: The length of the input sequence, including any special tokens.
    # c: The dimensionality of the hidden states, which is determined by
    #   the specific GPT-2 model variant. For example, the 'gpt2' model has a hidden size of 768.
    embeddings = outputs.last_hidden_state

# Access the embeddings for the first sequence (excluding first tokens)
first_token_embedding = embeddings[0, 1:]

# Print the embeddings
print(embeddings)
print(first_token_embedding)
```
